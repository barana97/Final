```python
# SECOND QUESTION Write a function that takes a list of integers and returns the smallest even element of that list.
# If there is no even number in the list, the function should return None. If the list contains any non-integers, 
#the function should print "Non-integer detected!" and continue with the next element of the list.

def smallest_even_element(lst):
    smallest_even = None

    for element in lst:
        if not isinstance(element, int):
            print("Non-integer detected!")
            continue

        if element % 2 == 0 and (smallest_even is None or element < smallest_even):
            smallest_even = element

    return smallest_even

# Example usage:
numbers = [3, 8, 5, 12, 'abc', 7, 4, 10, 'xyz']
result = smallest_even_element(numbers)

if result is not None:
    print(f"The smallest even element is: {result}")
else:
    print("No even number in the list.")
```

```python
# FOURTH QUESTION: (Write a function that takes a string,
#removes all commas and all words shorter than four characters from it, and returns the remaining string.)


def process_string(input_string):
    # Split the string into words
    words = input_string.split()

    # Remove commas and filter out words shorter than four characters
    filtered_words = [word.strip(",") for word in words if len(word) >= 4]

    # Join the filtered words into a new string
    result_string = " ".join(filtered_words)

    return result_string

# Example usage:
input_text = "Hello, this is a sample string with some short words, like hi, bye, and cat."
processed_text = process_string(input_text)
print("Original String:")
print(input_text)
print("\nProcessed String:")
print(processed_text)
```

```python
#QUESTION 6
#Read the files 'ATX.csv' and 'OMV.csv' that you can find in the folder Course Material / data into pandas 
#DataFrames, turning the Date column into a DatetimeIndex. Estimate rolling market betas 
#(where the ATX serves as the market index) for OMV by regressing OMV's daily returns on the daily returns of the ATX.#
#Use a rolling window of 252 days. Plot the estimates of beta over time.

# Read the data into pandas DataFrames
atx_data = pd.read_csv('path_to_data/ATX.csv', parse_dates=['Date'], index_col='Date')
omv_data = pd.read_csv('path_to_data/OMV.csv', parse_dates=['Date'], index_col='Date')

# Merge dataframes on Date
merged_data = pd.merge(omv_data, atx_data['Adj Close'], how='inner', left_index=True, right_index=True)
merged_data.columns = ['OMV_Adj_Close', 'ATX_Adj_Close']

# Calculate daily returns
merged_data['OMV_Returns'] = merged_data['OMV_Adj_Close'].pct_change()
merged_data['ATX_Returns'] = merged_data['ATX_Adj_Close'].pct_change()

# Define the rolling window size
rolling_window = 252

# Estimate rolling market betas
merged_data['Rolling_Betas'] = (
    merged_data['OMV_Returns'].rolling(window=rolling_window)
    .cov(merged_data['ATX_Returns'])
    / merged_data['ATX_Returns'].rolling(window=rolling_window).var()
)

# Plot the rolling betas over time
plt.figure(figsize=(12, 6))
plt.plot(merged_data['Rolling_Betas'], label='Rolling Betas')
plt.title('Rolling Market Betas for OMV with ATX')
plt.xlabel('Date')
plt.ylabel('Rolling Beta')
plt.legend()
plt.show()
```

```python
# QUESTION 7
#Read the file 'ATX.csv' that you can find in the folder Course Material / data into a pandas DataFrame, turning 
#the Date column into a DatetimeIndex. Drop all columns of the DataFrame except for  the Adjusted Close. 
#Change the frequency of the data from daily to monthly, using the values at the end of the month. 
#Compute and print the mean and the standard deviation of the monthly returns.

import pandas as pd

# Read the data into a pandas DataFrame
atx_data = pd.read_csv('path_to_data/ATX.csv', parse_dates=['Date'], index_col='Date')

# Keep only the 'Adj Close' column
atx_data = atx_data[['Adj Close']]

# Resample to monthly frequency using the last observation in the month
atx_monthly = atx_data.resample('M').last()

# Compute monthly returns
monthly_returns = atx_monthly.pct_change()

# Compute and print mean and standard deviation of monthly returns
mean_return = monthly_returns.mean().values[0]
std_dev_return = monthly_returns.std().values[0]

print(f'Mean of Monthly Returns: {mean_return:.6f}')
print(f'Standard Deviation of Monthly Returns: {std_dev_return:.6f}')
```

```python
#QUESTION 8 
#Read the file 'ATX.csv' that you can find in the folder Course Material / data into a pandas DataFrame. 
#Create a new column containing the absolute value of the daily relative price range defined as the difference between #
#the High and Low prices divided by the previous day's Adjusted Close. Create another column containing the absolute values of the daily returns. Make a scatter plot of the price range against the absolute values of the returns.

atx_data = pd.read_csv('path_to_data/ATX.csv', parse_dates=['Date'], index_col='Date')

# Create a new column for the absolute value of daily relative price range
atx_data['Price_Range'] = abs((atx_data['High'] - atx_data['Low']) / atx_data['Adj Close'].shift(1))

# Create another column for the absolute values of daily returns
atx_data['Daily_Returns'] = abs(atx_data['Adj Close'].pct_change())

# Scatter plot of Price Range against Daily Returns
plt.scatter(atx_data['Price_Range'], atx_data['Daily_Returns'], alpha=0.5)
plt.title('Scatter Plot: Price Range vs Daily Returns')
plt.xlabel('Price Range')
plt.ylabel('Daily Returns')
plt.show()
```
