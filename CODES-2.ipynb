```python
#This exercise asks you to build a simple portfolio management system.
#Write a Python class named Stock with attributes for stock symbol (a string), number of shares (an integer), and price (a float). Implement __init__ and __str__.
#Create a class named Portfolio that manages a collection of Stock objects. Implement methods to add a new stock, update an existing stock's information (e.g., change the number of shares or price), and delete a stock from the portfolio.
#Implement a method to read a portfolio from a text file. The file will have stock data in a comma-separated format (e.g., AAPL,50,130.10 for Apple stock).
#Implement a method to write the current state of the portfolio to a text file in the same format.
#Implement a method to calculate the total portfolio value using a list comprehension.
#Test your code.


class Stock:
    def __init__(self, symbol, shares, price):
        self.symbol = symbol
        self.shares = shares
        self.price = price

    def __str__(self):
        return f"Stock(symbol={self.symbol}, shares={self.shares}, price={self.price})"

    def update_shares(self, shares):
        self.shares = shares

    def update_price(self, price):
        self.price = price

class Portfolio:
    def __init__(self):
        self.stocks = {}

    def add_stock(self, stock):
        self.stocks[stock.symbol] = stock

    def update_stock(self, symbol, shares=None, price=None):
        if symbol in self.stocks:
            if shares is not None:
                self.stocks[symbol].update_shares(shares)
            if price is not None:
                self.stocks[symbol].update_price(price)

    def delete_stock(self, symbol):
        if symbol in self.stocks:
            del self.stocks[symbol]

    def read_from_file(self, filename):
        with open(filename, 'r') as file:
            for line in file:
                symbol, shares, price = line.strip().split(',')
                self.add_stock(Stock(symbol, int(shares), float(price)))

    def write_to_file(self, filename):
        with open(filename, 'w') as file:
            for stock in self.stocks.values():
                file.write(f"{stock.symbol},{stock.shares},{stock.price}\n")

    def total_investment(self):
        return sum(stock.shares * stock.price for stock in self.stocks.values())
    
    
test_file_name = 'portfolio.txt'
sample_data = """AAPL,50,130.10
GOOG,30,1500.5
MSFT,20,200.40"""

with open(test_file_name, 'w') as file:
    file.write(sample_data)

portfolio = Portfolio()
portfolio.read_from_file(test_file_name)
portfolio.add_stock(Stock("TSLA", 10, 600.00))
portfolio.update_stock("AAPL", shares=60)
portfolio.delete_stock("GOOG")
portfolio.write_to_file(test_file_name)

print("Total Investment:", portfolio.total_investment())    
        
```

```python
#Create a 2D array where each of 5 rows represents random weekly prices of different stocks over a year (52 weeks), 
#generated from a starting value of 100 and weekly logarithmic returns with mean 0.0025 and standard deviation of 0.05. Plot the price path.
#Split the data into four quarters and calculate and print the average price 
#as well as the lowest and highest price of each stock for each quarter.

import numpy as np
import matplotlib.pyplot as plt

num_stocks = 5
num_weeks = 52
starting_value = 100
mean_return = 0.0025
std_dev_return = 0.05

rng = np.random.default_rng(42)

log_returns = rng.normal(mean_return, std_dev_return, (num_stocks, num_weeks))
prices = starting_value * np.exp(np.cumsum(log_returns, axis=1))



def quarter_stats(quarter_prices):
    avg_prices = np.mean(quarter_prices, axis=1)
    min_prices = np.min(quarter_prices, axis=1)
    max_prices = np.max(quarter_prices, axis=1)
    return avg_prices, min_prices, max_prices

quarterly_prices = np.array_split(prices, 4, axis=1)

for i, quarter_prices in enumerate(quarterly_prices, start=1):
    avg_prices, min_prices, max_prices = quarter_stats(quarter_prices)
    print(f"Quarter {i}:")
    for stock_index in range(num_stocks):
        print(f"  Stock {stock_index+1}: Average Price: {avg_prices[stock_index]:.2f}, "
              f"Lowest Price: {min_prices[stock_index]:.2f}, Highest Price: {max_prices[stock_index]:.2f}")
        
        
plt.figure(figsize=(14, 7))
for i in range(num_stocks):
    plt.plot(prices[i], label=f'Stock {i+1}')
plt.title('Weekly Prices of 5 Stocks Over a Year (Log Returns) with RNG')
plt.xlabel('Weeks')
plt.ylabel('Price')
plt.legend()
plt.show()        
```

```python
#Predict the next month's change in interest rates (10-Year Treasury Constant Maturity Rate) using current economic indicators and their changes. For this purpose, download the following data for the period 2000 to 2023 from FRED via pandas_datareader:
#10-Year Treasury Constant Maturity Rate (DGS10)
#Consumer Price Index for All Urban Consumers: All Items (CPIAUCSL)
#Civilian Unemployment Rate (UNRATE)
#Combine the data in one dataframe of monthly frequency, using the last available data within each month.
#For the CPI, use the percentage change instead of the level of the time series.
#Create new columns containing the monthly differences of the three time series.
#Run an OLS regression with the next month's change in interest rates as the dependent variable and the levels and changes in the unemployment rate and inflation as the independent variables.

import pandas as pd
import pandas_datareader.data as web
import statsmodels.api as sm
from datetime import datetime

# Define the time frame for the data
start = datetime(2000, 1, 1)
end = datetime(2023, 12, 31)

# Fetching the data from FRED
dgs10 = web.DataReader('DGS10', 'fred', start, end) # 10-Year Treasury Rate
cpiaucsl = web.DataReader('CPIAUCSL', 'fred', start, end) # CPI for All Urban Consumers
unrate = web.DataReader('UNRATE', 'fred', start, end) # Unemployment Rate

# Merging the datasets
data = pd.concat([dgs10, cpiaucsl, unrate], axis=1)
data.columns = ['DGS10', 'CPI', 'Unemployment']
data = data.resample('M').last()

data['CPI_pct_change'] = data['CPI'].pct_change() * 100  # Convert to percentage

# Calculating month-over-month changes for the independent variables
data['CPI_pct_change_diff'] = data['CPI_pct_change'].diff()
data['Unemployment_change'] = data['Unemployment'].diff()

# Shifting the dependent variable (interest rate) to predict the next month's change
data['DGS10_next_month_change'] = data['DGS10'].diff().shift(-1)

# Dropping initial rows with NaN values due to differencing and the last row due to shift
data = data.dropna()
data.head()


# Setting up the regression model
Y = data['DGS10_next_month_change'] # Dependent variable
X = data[['CPI_pct_change', 'Unemployment', 'CPI_pct_change_diff', 'Unemployment_change']] # Independent variables
X = sm.add_constant(X) # Adding a constant term to the predictor

# Performing the regression
model = sm.OLS(Y, X).fit()

# Outputting the summary of the regression
model.summary()




robust_model = model.get_robustcov_results(cov_type='HC0')
robust_model.summary()



# Choosing the lag. You can specify a number or use an automatic selection criterion
# For example, let's use 4 lags
lag = 4

# Using Newey-West standard errors
newey_west_model = model.get_robustcov_results(cov_type='HAC', maxlags=lag)

# Outputting the summary of the regression with Newey-West standard errors
print(newey_west_model.summary())
```
